```python
#Step 0. 필요한 모듈과 라이브러리를 로딩합니다.
import sys
import os
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
from selenium import webdriver
import time
from tqdm import tqdm_notebook
```

```python
# 크롬창 띄우기
driver = webdriver.Chrome("chromedriver.exe")  # 크롬 드라이버 로드
driver.get("https://www.coupang.com/")  # "#차량용방향제" 검색
time.sleep(3)
```

```python
query_text = "곰돌이푸우인형"
```

```python
# 검색 창에 검색어 입력 후 검색
element = driver.find_element_by_css_selector('#headerSearchKeyword')

element.send_keys(query_text)  # query_txt는 위에서 입력한 키워드
element.submit()
time.sleep(1)

# 판매량순 클릭
driver.find_element_by_xpath("""//*[@id="searchSortingOrder"]/ul/li[4]/label""").click( ) 
```

```python
# url 수집
# elements : 여러개
# element : 하나
urls = driver.find_elements_by_css_selector('.search-product-link')
urls[0]
```

```python
urls[0].get_attribute('href')
```

```python

url_list=[]
for url in urls:
    url = url.get_attribute('href')
    url_list.append(url)
    
url_list
```

```python
tmp = pd.DataFrame({'url' : url_list})  #이차원 데이터프레임으로 바꿔줌
tmp       
```

```python
# 저장하기
tmp.to_excel('coupang_url.xlsx')
```

# 쿠팡 리뷰 블로그 내용 크롤링하기

```python
# "url_list.csv" 불러오기
url_load = pd.read_excel("coupang_url.xlsx")
num_list = len(url_load)

print(num_list)
url_load
```

```python
print(num_list)
```

```python
save = pd.DataFrame({'product': [],'user' : [], 'rating' : [], 'review' : [] , 'date' : []})

for n in range(0,8) :
    # 글 띄우기
    url = url_load['url'][n]
    driver = webdriver.Chrome("chromedriver.exe")  # 윈도우는 "chromedriver.exe"
    driver.get(url)   # 글 띄우기

    # 제품명 크롤링
    product = driver.find_element_by_css_selector('.prod-buy-header__title').text
    print(product)
    time.sleep(1)
    
    # 상품평 클릭
    try:  # 상품평이 있을 경우
        driver.find_element_by_css_selector('.count').click()
        time.sleep(1)
    except: # 상품평이 없을 경우
        time.sleep(1)
        
    try:
        i=1
        user_list = []
        rating_list = []
        review_list = []
        date_list = []
        while i :
            # user 수집
            users = driver.find_elements_by_css_selector('.sdp-review__article__list__info__user__name.js_reviewUserProfileImage')
            for user in users:
                user = user.text
                user_list.append(user)
            # rating 수집
            ratings = driver.find_elements_by_css_selector('.sdp-review__article__list__info__product-info__star-orange.js_reviewArticleRatingValue')

            for rating in ratings:
                rating = int(rating.get_attribute('data-rating'))
                rating_list.append(rating)

            # review 수집
            review_box = driver.find_elements_by_css_selector('.sdp-review__article__list.js_reviewArticleReviewList')

            for k in range(len(review_box)) :
                try:
                    reviews = review_box[k].find_element_by_css_selector('.sdp-review__article__list__review__content.js_reviewArticleContent')
                    review = reviews.text
                except:
                    review = ""
                review_list.append(review)

            # date 수집
            dates = driver.find_elements_by_css_selector('.sdp-review__article__list__info__product-info__reg-date')

            for date in dates:
                date = date.text
                date_list.append(date)

            #다음 페이지 넘어가기
            i = i+1
            if i%12!= 0 :
                driver.find_element_by_xpath(""" //*[@id="btfTab"]/ul[2]/li[2]/div/div[5]/section[4]/div[3]/button["""+(str)(i%12+1) +"""]""").click( ) 

            time.sleep(1)

    except:
            if len(user_list) == 0 :
                user_list.append("None")
                rating_list.append("None")
                review_list.append("None")
                date_list.append("None")
            tmp = pd.DataFrame({'product': product,'user' : user_list, 'rating' : rating_list, 'review' : review_list, 'date' : date_list})
            save = pd.concat([save,tmp])

   
```
